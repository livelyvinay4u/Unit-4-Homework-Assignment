```python
# initial imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
```

```python
# drop null values in whale returns DataFrame and verify null values deleted
df_whale_modified = df_whale.dropna()
df_whale_modified.isnull().sum()
```

```python
df_whale_modified
```

```python
#remove $ and , from sptsx Close cloumn values
df_sptsx["Close"] = df_sptsx["Close"].replace('[$,]','',regex=True)
#print(df_sptsx.dtypes)
df_sptsx
```

```python
#Convert Close coloumn dtype to float
df_sptsx["Close"] = df_sptsx["Close"].astype(float)
print(df_sptsx.dtypes)
```

```python
#df_whale_modified.iloc[[0],:]
```

```python
# Drop duplicate records in whale returns DataFrame
df_whale_modified = df_whale_modified.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
# Check if duplicates exist in whale returns DataFrame
df_whale_modified.duplicated()
```

```python
# Check how many rows and columns are there in each DataFrame after cleaning
df_whale_modified.shape,df_algo.shape,df_sptsx.shape
```

```python
#Convert S&P TSX 60 Returns closing price to daily returns
sptsx_daily_returns = df_sptsx.pct_change()
sptsx_daily_returns.head()
```

```python
#Join `Whale Returns`, `Algorithmic Returns`, and the `S&P TSX 60 Returns` into a single DataFrame
joined_data_rows = pd.concat([df_whale_modified, df_algo, sptsx_daily_returns], axis = 1, join = "inner")
joined_data_rows
```

```python
# Reset index of joined DataFrame
joined_data_rows.reset_index(inplace=True)
```

```python
# Select records in the joined DataFrame for common date range in whale returns, algo returns and sptsx history DataFrames
joined_data_rows = joined_data_rows[(joined_data_rows['Date'] >= '2015-03-03') & 
                                    (joined_data_rows['Date'] <= '2019-05-01')]
joined_data_rows.head()
```

```python
# Set the index back to Date, of joined DataFrame
joined_data_rows = joined_data_rows.set_index("Date")
joined_data_rows.head()
```

```python
# Caclulate cummulative returns of all portfolios
cumulative_returns = (1 + joined_data_rows).cumprod()
cumulative_returns.head()
```

```python
# calculate Standard Deviation for whale returns portfolio
whale_std = df_whale_modified.std()
whale_std = whale_std.sort_values(ascending=False)
whale_std
```

```python
# calculate Standard Deviation for algo returns portfolio
algo_std = df_algo.std()
algo_std = algo_std.sort_values(ascending=False)
algo_std
```

```python
# calculate Standard Deviation for sp tsx 60 history
sptsx_std = sptsx_daily_returns.std()
sptsx_std
```

```python
# Calculate standard deviation for all the portfolios including market
joined_std = joined_data_rows.std()
joined_std = joined_std.sort_values(ascending=False)
joined_std
```

```python
# calculate annualized Standard Deviation
joined_annualized_std = joined_std * np.sqrt(252)
joined_annualized_std
```

```python
#Calculate rolling standard deviation for all portfolios using a 21-day window.
joined_rolling = joined_data_rows.rolling(window=21).std()
joined_rolling
```

```python
#variance 
df_sptsx_varince = joined_data_rows['Close'].var()
df_sptsx_varince
```

```python
#co-variance
df_sptsx_covarince = joined_data_rows['Algo 1'].cov(joined_data_rows['Close'])
df_sptsx_covarince
```

```python
#Calculate beta
algo1_beta = df_sptsx_covarince/df_sptsx_varince
algo1_beta
```

```python
# Calculate 30-day rolling covariance of AMZN vs. S&P 500 and plot the data
rolling_covariance = joined_data_rows['Algo 1'].rolling(window=60).cov(joined_data_rows['Close'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 60-Day Covariance of Algo1 Returns vs. S&P TSX 60 Returns')
```
